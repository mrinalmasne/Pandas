```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('imdb-top-1000.csv')
```

```python
movies.head()
```

```python
genres=movies.groupby('Genre')
```

```python
#Applying builtin aggregation functions on groupby objects
genres.sum()
```

```python
#find the top 3 genres by total earnings
movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
#find director with most popularity
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
#Groupby Attributes and Methods
#find total number of groups-->len
len(movies.groupby('Genre'))
```

```python
#find items in each group-->size
movies.groupby('Genre').size()
```

```python
#first() /last(0-->nth item
genres=movies.groupby('Genre')
genres.first()
```

```python
genres.last()
```

```python
genres.nth(5)
```

```python
#get_group-->vs filtering
genres.get_group('Horror')
```

```python
#filtering--slower than set_group
movies[movies['Genre']=='Fantasy']
```

```python
#groups attribute
genres.groups
```

```python
#describe
genres.describe()
```

```python
#sample
genres.sample()
```

```python
genres.nunique()
```

```python
#agg method---used to apply different aggregate functions on different columns
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
    }
        

)
```

```python
#looping on groups
for group,data in genres:
    print(data)
```

```python
#split (apply) combine
#find number of movies starting with A for each group
def foo(group):
    return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(foo)
```

```python
#group by on multiple column
duo=movies.groupby(['Director','Star1'])
duo
duo.size()
```
